```python
from pyspark import SparkConf, SparkContext
import collections

conf = SparkConf().setMaster('local').setAppName('RatingsHistogram')
sc = SparkContext.getOrCreate(conf = conf)
rdd=sc.textFile('adventures.txt')
```

```python
result=rdd.collect()
print(result)
```

```python
for value in result:
    print(value)
```

```python
def Func(lines):
    lines=lines.lower()
    lines=lines.split()
    return lines
    
 
```

```python
rdd_func=rdd.map(Func)
rdd_func.take(10)
```

```python
rdd2=rdd.flatMap(Func)
rdd2.take(10)
```

```python
rdd3=rdd2.map(lambda x:(x,1))
rdd3.take(10)
```

```python
rdd4=rdd3.reduceByKey(lambda a,b:a+b)
rdd4.take(10)
```

```python
rdd5=rdd4.sortByKey()
rdd5.take(10)
```

```python
rdd6=rdd4.map(lambda x:(x[1],x[0])).sortByKey()
rdd6.take(500)
```

```python
rdd7=rdd6.filter(lambda x: 'a' in x[1])
rdd7.take(10)
```

```python
rdd7_1=rdd6.filter(lambda x:(x[0] % 2) ==0 )
rdd7_1.take(10)
```

```python
rdd8=rdd3.groupBy(lambda x:x[0:3])
#rdd8.take(1)
print([(k,list(v)) for (k,v) in rdd8.take(1)])
#rdd8.collect()
```

```python
#Perform a task(eg. count words) separately  on each parttion and get the output of the task performed in these partitions
#works like ma but runs separately on different partitions of a RDD
def func(iterator):
    count_SHERLOCK=0
    count_HOLMES=0
    for i in iterator:
        if i=='SHERLOCK':
            count_SHERLOCK=count_SHERLOCK+1
        if i=='HOLMES':
            count_HOLMES=count_HOLMES+1
    return(count_SHERLOCK,count_HOLMES)
```

```python
rdd9=rdd2.mapPartitions(func).take(10)

```

```python
rdd10=rdd8.sample(False,0.4)
rdd10.collect()
```

```python
sample1=rdd8.sample(False,0.2)
sample2=rdd8.sample(False,0.2)
union_rdd=sample1.union(sample2)
union_rdd.collect()
```

```python
sample1=rdd8.sample(False,0.2)
sample2=rdd8.sample(False,0.2)
join_rdd=sample1.join(sample2)
join_rdd.collect()
```
